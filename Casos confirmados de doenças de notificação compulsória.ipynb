```python
#Começaremos o trabalho de limpeza dos dados, 
#primeiro começaremos convertendo a coluna ano de inteira para o tipo data


dados['ano'] = dados['ano'].astype(int)

# Verificando o tipo das colunas após a conversão
print(dados.dtypes)

```

```python
#Selecionarei as colunas com as quais iremos trabalhar

colunas_desejadas = ['município', ' doençasdenotificaçãocompulsória', 'ano','Quantidadesdedoençasnotificados ']

dados = dados[colunas_desejadas]

dados.head()
```

```python
#Agora excluiria as linhas de dados que com erro 

linhas_com_zero = dados[dados.eq(0).any(axis=1)]
dados_sem_zero = dados.drop(index=linhas_com_zero.index)

```

```python
print(dados_sem_zero)
```

```python
dados.head()
```

```python
dados.isna()
```

```python
#vamos nos livrar desses erros de municipio

linhas_com_false = dados[dados.apply(lambda row: not all(row), axis=1)]
```

```python
dados_sem_false = dados.drop(index=linhas_com_false.index)

print(dados_sem_false)
```
